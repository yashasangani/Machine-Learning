```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/yashasangani/Datasets/main/cars.csv",delimiter=",")
df.head()
```

```python
df["owner"].value_counts()
```

# 1.One Hot Encoding using PANDAS

```python
pd.get_dummies(df,columns = ['fuel','owner'],dtype= 'int64')
```

# 2. K-1 One Hot Encoding

```python
pd.get_dummies(df,columns=["fuel","owner"],drop_first=True,dtype= 'int64')
```

# 3. OneHotEncoding using Sklearn

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder(drop ='first',sparse = False,dtype=np.int32)
x_train_new=oe.fit_transform(x_train[['fuel','owner']])
x_test_new = oe.transform(x_test[['fuel','owner']])
```
